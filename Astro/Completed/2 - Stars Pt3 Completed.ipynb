## Magnitudes

We use "magnitudes" to describe how bright a star is. Remember, we mentioned that stars give off a lot of light. That light makes it to Earth, and we can measure how bright the star is from that light. The magnitude system was developed a long time ago, and unforunately, it is a weird scale. The higher a star's magnitude, the DIMMER it is. Bright stars have a low magnitude. (This is backwards from how I would personally do it! I try to remember this scale by saying the magnitude is the number of telescopes you'd need to see the star. If you need a high number, it must be dim.)

There's another thing that is special about star light. It is a lot of different kinds of light! Humans can see the "visible light," and we feel what's called "infrared light" has heat. While our eyes can only process visible light, we can use machines to "see" the other kind of light coming from a star. We use "filters" to limit the light to a certain kind. This is similar to how wearing sunglasses block out light!

## Color

Color is another thing astronomers define a little weirdly. We are used to seeing a color that our eyes can pick up. That color corresponds to a certain kind of light.

Astronomers define color by subtracting the magnitudes in two different filters. For example, we could do g - r. That will give us a number, not what most people would call a color. We can say colors that are bigger than 0.6 are redder. And those that are smaller than 0.6 are bluer.

```python
# calculate color and save the values to a variable
color = g - r
```

```python
# create a histogram of color
# add a vertical line at 0.6
# label the left side "bluer" and the right side "redder" using plt.annotate('label', (x,y), color='color')

plt.hist(color)
plt.axvline(0.6, color='black')
plt.annotate('bluer', (0.15, 12000), color='blue', fontsize=16)
plt.annotate('redder', (1.3, 12000), color='red', fontsize=16)
plt.xlabel('Color [g-r]')
plt.ylabel('Number of Stars')
```

```python
# model fitting for color vs temperature 
# let's see if an inversely proportional model actually describes the data well

# import our model fitting package
from scipy.optimize import curve_fit
import numpy as np

# create a function for an inversely proportional model
def inversely_proportional(x, A):
    y = A / x # A is the inversely proportional constant
    return y

# apply the model to our data
popt, pcov = curve_fit(inversely_proportional, temperature, color) # replace x and y with what is on your axes above

# create model data
x_model = np.linspace(3000, 12000, 1000)
y_model = inversely_proportional(x_model, popt)

# copy and paste the code you had above for your color vs temperature graph
# use plt.plot to plot the (x_model, y_model) data
# label what is the data and what is the model
plt.scatter(temperature, color, label='Data')
plt.axhline(0.6, color='black')
plt.annotate('bluer', (4000, .25), color='blue', fontsize=16)
plt.annotate('redder', (9000, .75), color='red', fontsize=16)
plt.ylabel('Color [g-r]')
plt.xlabel('Temperature [K]')
plt.plot(x_model, y_model, color='black', label='Model', linewidth=2, linestyle='--')
plt.legend()
```

**Does the model appear to be a good fit? Would it have a high or low RMSE? Is the relationship between color and temperature a perfect inversely proportional fit?**

**Answer:** The model is not a perfect fit. It wouldn't have the hight RMSE but not the lowest either. This means the relationship between color and temperature is not perfectly inversely proportional.
